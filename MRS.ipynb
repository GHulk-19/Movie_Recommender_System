```python
import numpy as np
import pandas as pd
import ast
import nltk
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
#credits.head(1)['cast'].values
credits.head(1)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
# Columns required "genre, id(to fetch posters)/movie_id, keywords, title, overview, cast, crew"
movies=movies[['movie_id','title','overview','genres', 'keywords', 'cast','crew']]
```

```python
movies.head()
```

```python
#checking for missing data
#3 rows has overview missing
movies.isnull().sum()
```

```python
#drop the missing data
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#string of list to be converted to list
import ast
```

```python
#[{'id': 28, 'name': 'Action'},
 #{'id': 12, 'name': 'Adventure'},
 #{'id': 14, 'name': 'Fantasy'},
 #{'id': 878, 'name': 'Science Fiction'}]
# convert to ['Action', 'Adventure', 'Fantasy', "Science Fiction"]

def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['genres'].head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['keywords'].head()
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python

def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            l.append(i['name'])
    return l
    
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# overview is a string and others in list form... now we cannot merge all the other columns.. so we can convert it to list using lambda function
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# in cast "sam worthington" there is a space ... we must remove it to make a single string
#Becoz if we dont remove..."Sam" and "Worthington" becomes two entites...but both are name of one person
# Convert it as "SamWorthington"
# apply to any column if space exsists

movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
# create a column "tags".. that stores the concatanation of overview,keywords,crew,cast,genres.
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#we can remove the other columns or create a new dataframe
```

```python
#creating new dataframe
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
# convert the tags list into string ... every element in list is separated by space
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))

```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
#stemming words
#['learning,learnt,learns,learn] ---> [light,light,light,light]
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

##  Recommendation ##

```python
new_df[new_df['title']=="Avatar"].index[0]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    return
```

```python
recommend("Batman Begins")
```
